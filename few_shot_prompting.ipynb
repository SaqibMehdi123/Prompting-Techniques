```python
from transformers import pipeline

gen = pipeline('text-generation', model='EleutherAI/gpt-neo-125M', device=0)
```

```python
def few_shot(text):
  return f"""Classify the sentiment.

  Text: I loved the movie!
  Sentiment: positive

  Text: I hated the food.
  Sentiment: negative

  Text: The service was great.
  Sentiment: positive

  Text: It was a horrible experience.
  Sentiment: negative

  Text: This phone is amazing.
  Sentiment: positive

  Text: The room smelled bad.
  Sentiment: negative

  Text: {text}
  Sentiment:"""

def zero_shot(text):
  return f"Classify the sentiment of this sentence as positive or negative:\n\n{text}\nSentiment:"
```

```python
from termcolor import colored

samples = [
    {"text": "I hate waiting in lines.", "label": "negative"},
    {"text": "The experience was amazing!", "label": "positive"},
    {"text": "The movie was a waste of time.", "label": "negative"},
    {"text": "Absolutely love this app!", "label": "positive"},
    {"text": "It was okay, not great.", "label": "negative"},  # slightly ambiguous
    {"text": "This is the best purchase I've made.", "label": "positive"},
    {"text": "The delivery was late and the item was broken.", "label": "negative"},
]

def test(mode):
  print(f"\n{'='*40}\n🔍 Sentiment Classification ({mode.upper()} MODE)\n{'='*40}")
  correct = 0

  for idx, s in enumerate(samples, 1):
    prompt = few_shot(s['text'] if mode=='fewshot' else zero_shot(s['text']))
    out = gen(prompt, max_new_tokens=10, do_sample=False)[0]['generated_text']
    # print only the part of the raw output
    print(f"Raw output: {out.split('Text:')[-1].strip()}")
    sentiment_part = out.split('Sentiment:')[-1].strip()
    if sentiment_part:
        pred = sentiment_part.split()[0].lower()
    else:
        pred = "none"

    expected = s['label'].lower()
    is_correct = pred.startswith(expected)
    result_tag = '✅ Correct' if is_correct else '❌ Incorrect'

    print(f"\n🔹 Example # {idx}")
    print(f"📝 Text       : {s['text']}")
    print(f"🎯 Expected   : {expected}")
    print(f"🤖 Predicted  : {pred}")
    print(f"📊 Result     : {colored(result_tag, 'green' if is_correct else 'red')}")

    correct += is_correct

  total = len(samples)
  accuracy = correct / total
  print(f'\n✅ Final Accuracy ({mode.upper()}): {correct}/{total} -> {accuracy:.2f}')
```

```python
test("fewshot")
test('zeroshot')
```
